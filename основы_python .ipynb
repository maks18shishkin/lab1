#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
# Your code goes here
def func1(string):
    abc = 'aeiou'
    cnt = 0
    for i in string.lower():
        if i in abc:
            cnt += 1
    return cnt
test_list = [
    'qwerty',
    'aaabbbccc',
    'pooooptptptptpgpgpfpgfdgkl;sdfagiasgjafhgharghareljkgha;']

for i in test_list:
    print(func1(i))
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def calculate_steps(value):
    counter = 0
    while value > 10:
        digits = map(int, list(str(value)))
        product = 1
        for digit in digits: product *= digit
        value = product
        counter += 1
    return counter

test_data = [39, 40, 10, 12]
for num in test_data:
    print(calculate_steps(num))
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def calculate_mse(vector1, vector2):
    total = 0
    for a, b in zip(vector1, vector2):
        total += (a - b) ** 2
    return total / len(vector1)

print(calculate_mse([1, 2, 3], [2, 3, 4]))
print(calculate_mse([1, 3, 3], [2, 9, 4]))
print(calculate_mse([6, 2, 4], [2, 3, 2]))
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
# Your code goes here
def check_prime(num):
    if num < 2:
        return False
    for divisor in range(2, int(num**0.5) + 1):
        if num % divisor == 0:
            return False
    return True

def factorize_number(number):
    prime_numbers = [i for i in range(2, number + 1) if check_prime(i)]
    factorization = []
    current_value = number
    
    for prime in prime_numbers:
        if prime * prime > current_value:
            break
        while current_value % prime == 0:
            factorization.append(prime)
            current_value //= prime
    
    if current_value > 1:
        factorization.append(current_value)
    
    factor_count = {}
    for factor in factorization:
        factor_count[factor] = factor_count.get(factor, 0) + 1
    
    output_string = ''
    for prime_factor in sorted(factor_count.keys()):
        exponent = factor_count[prime_factor]
        output_string += f'({prime_factor}**{exponent})'
    
    return output_string
test_cases = [
    86240,
    2*2*2*2*11*11,
    2*2*2,
    39856,
    737735
]

for test_value in test_cases:
    print(factorize_number(test_value))
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def check_pyramid_possible(total_cubes):
    current_sum = 1
    level = 2
    
    while total_cubes >= current_sum:
        current_sum += level ** 2
        level += 1
    
    if total_cubes == (current_sum - (level-1)**2):
        return level - 2
    else:
        return 'It is impossible'
test_cases = [
    1 + 2*2 + 3**2 + 4**2 + 5**2,  # 55
    2,
    6,
    18,
    1 + 2*2 + 3**2 + 4**2 + 5**2 + 6**2 + 7**2  # 140
]

for cubes in test_cases:
    result = check_pyramid_possible(cubes)
    print(result, cubes)
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
# Your code goes here
def is_number_balanced(num_value):
    num_string = str(num_value)
    str_length = len(num_string)
    
    if str_length % 2 == 0:
        left_part_end = str_length // 2 - 1
        right_part_start = str_length // 2 + 1
        sum_before_middle = sum(int(char) for char in num_string[:left_part_end])
        sum_after_middle = sum(int(char) for char in num_string[right_part_start:])
        return sum_before_middle == sum_after_middle
    else:
        middle_position = str_length // 2
        sum_left_side = sum(int(char) for char in num_string[:middle_position])
        sum_right_side = sum(int(char) for char in num_string[middle_position + 1:])
        return sum_left_side == sum_right_side
test_values = [
    '11211',
    '112211',
    '1' * 1000 + '2' + '1' * 1000,
    '1' * 1000 + '22' + '1' * 1000,
    '65736573657650235067' * 1000
]

for test_num in test_values:
    result = is_number_balanced(test_num)
    print(result)
```
